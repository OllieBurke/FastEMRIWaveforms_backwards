```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.chdir("../data_files")
```

```python
a_vec = np.load("a_vec.npy")
e0_vec = np.load("e0_vec.npy")
```

```python
SNR_AAK_vec_a_against_SNR = [np.load("SNR_AAK_vec_a_e0_0p{}.npy".format(i)) for i in range(1,6)]
SNR_Kerr_vec_a_against_SNR = [np.load("SNR_Kerr_vec_a_e0_0p{}.npy".format(i)) for i in range(1,6)]
```

```python
e_plot_vec = [i/10 for i in range(1,6)]
i = 4
plt.plot(a_vec,SNR_Kerr_vec_a_against_SNR[i], label = 'Kerr amplitudes')
plt.plot(a_vec,SNR_AAK_vec_a_against_SNR[i], label = 'AAK amplitudes')
plt.grid()
plt.xlabel(r'Spin parameter $a$')
plt.ylabel(r'SNR')
plt.title("Strong field (plunge): (M,mu,a, T) = (1e6, 10, {}, 2 years)".format(e_plot_vec[i]))
plt.legend()
# plt.savefig("SNR_plots/Eccentricity_Plot_a_0p7_strong_field_nmodes_AAK_50_specific_modes_kerr_None.pdf",bbox_inches = 'tight')
plt.show()
```

```python
SNR_AAK_vec_e_against_SNR = [np.load("SNR_AAK_vec_a_0p{}.npy".format(i)) for i in range(1,9,2)]
SNR_Kerr_vec_e_against_SNR = [np.load("SNR_Kerr_vec_a_0p{}.npy".format(i)) for i in range(1,9,2)]
```
